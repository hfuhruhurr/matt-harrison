```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

### Read in data

```python
auto = pd.read_csv('./vehicles.csv', low_memory=False)
```

### Make life simple

```python
makes = ['Chevrolet', 'Ford', 'Dodge', 'GMC', 'Toyota']
years = [1985, 1990, 1995, 2000]
```

### Render a Seaborn plot

```python
fig, ax = plt.subplots(figsize=(6,4))
sns.countplot(x='make', data=auto, ax=ax, order=makes, hue='year', hue_order=years)
```

### Do the same thing but in `pandas`

```python
(pd.crosstab(auto.make, auto.year) 
    .loc[makes, years]
   .plot.bar()
)
```

### Do the same thing again but different in `pandas`

```python
(auto 
    .groupby(['year', 'make'])
    .size()
    .unstack(level='year')
    .loc[makes, years]
    .plot.bar(figsize=(12,2))
)
```
